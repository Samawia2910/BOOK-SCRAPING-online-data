```python
from selenium import webdriver
from bs4 import BeautifulSoup
from selenium.common.exceptions import TimeoutException
from selenium.webdriver.chrome.options import Options 
import pandas as pd


chrome_options=Options()
prefs = {"profile.managed_default_content_settings.images":2}
chrome_options.add_experimental_option("prefs",prefs)
driver = webdriver.Chrome(executable_path = r"E:\BSCS\Semester 3\DSA\chromedriver.exe",options=chrome_options)
Name = []
Author = []
Price = []
Category =[]
Publisher = []
Isbn = []
Pages = []

pageNum=2182
while True:
    try:
        driver.get(f"https://www.readings.com.pk/pages/category.aspx?Category=&Level=&Sortby=ArrivalDate&BookType=&Page={pageNum}")
        pageNum += 1
        content = driver.page_source
        soup = BeautifulSoup(content)

        details = soup.findAll('div',attrs={'class':'product_detail_page_outer_colum'})
        if len(details)==0:
            break
        for i in range(len(details)):
            book = details[i]
            bookHeading = book.find('h5')
            name = bookHeading.find('a').text 
            Name.append(name)

            authorHeading = book.find('h6')
            author = authorHeading.find('a').text.strip()
            Author.append(author)

            priceHeading = book.find("div", attrs = {"class":'list_price'})
            price = priceHeading.findAll('span')
            Price.append(price[2].text.strip())

            categoryHeading = book.find("div", attrs = {"class": "books_category"})
            category = categoryHeading.find('a').text.strip()
            Category.append(category)

            publisherHeading = book.find("div", attrs = {"class": "books_publisher"})
            heading = publisherHeading.findAll('span')
            publisher = heading[0].text[:-1].strip()
            isbn = heading[1].text[:-2].strip()
            pages = heading[2].text.strip()
            Publisher.append(publisher)
            Isbn.append(isbn)
            Pages.append(pages)

            dp = pd.DataFrame({
                'Name':Name,
                'Author':Author,
                'Price':Price,
                'Category':Category,
                'Publisher':Publisher,
                'Isbn':Isbn,
                'Pages':Pages
            })
            
            dp.to_csv('Data2.csv',index=False)
        print(pageNum)
    except (TimeoutException):
        driver.refresh()
        sleep(6)
    
driver.close()
```
